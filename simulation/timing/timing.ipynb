```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_steps = [10, 100, 1000, 5000]
n_steps = n_steps * 3
n_steps.sort()
n_steps
```

```python
n_steps = [10, 10, 10, 100, 100, 100, 1000, 1000, 1000, 5000, 5000, 5000]
c_times = [0.00444987, 0.00646971, 0.00450458, 0.0662379, 0.0346345, 0.0331742, 0.388556, 0.409026, 0.382364, 1.80996, 1.81044, 1.81]
python_times = [0.005930423737,
0.0166926384,
0.006109237671,
0.04570198059,
0.04669523239,
0.0452747345,
0.4494979382,
0.4599728584,
0.4626333714,
2.306562662,
2.260612965,
2.253498554,]
```

```python
n_steps = np.array(n_steps)
c_times = np.array(c_times)
python_times = np.array(python_times)
```

```python
# Unique n_steps
unique_n_steps = np.unique(n_steps)

# Compute mean and std for each group
c_means = [np.mean(c_times[n_steps == n]) for n in unique_n_steps]
c_stds = [np.std(c_times[n_steps == n]) for n in unique_n_steps]
python_means = [np.mean(python_times[n_steps == n]) for n in unique_n_steps]
python_stds = [np.std(python_times[n_steps == n]) for n in unique_n_steps]

# Bar width
bar_width = 0.35
x = np.arange(len(unique_n_steps))  # x positions

# Plot
plt.figure(figsize=(10, 6))
plt.bar(x - bar_width / 2, c_means, bar_width, label='C++', yerr=c_stds, capsize=5)
plt.bar(x + bar_width / 2, python_means, bar_width, label='Python', yerr=python_stds, capsize=5)

# Labels and legend
plt.xlabel('Number of Steps (n)', fontsize=12)
plt.ylabel('Time (seconds)', fontsize=12)
plt.title('C++ vs Python Timings', fontsize=14)
plt.xticks(x, unique_n_steps)
plt.legend()
plt.tight_layout()
plt.show()
```
